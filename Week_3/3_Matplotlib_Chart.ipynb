# 차트의 요소

## 1. Text API in Matplotlib

````python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
````

바꿀 수 있는 요소들  

|family|style|variant|weight|size|  
|:---:|:---:|:---:|:---:|:---:|
|serif|normal|normal|light|xx-small|
|sans-serif|italic|small_caps|normal|x-small|
|cursive|oblique|---|medium|small|
|fantasy|---|---|semibold|medium|
|monospace|---|---|bold|large|
|---|---|---|heavy|x-large|
|---|---|---|black|xx-large|

다른 커스텀할 수 있는 요소들
  * ```color```
  * ```linespacing```
  * ```backgroundcolor```
  * ```alpha```
  * ```zorder```
  * ```visible```
  
정렬 관련 요소
  * ```va``` : vertical align
  * ```ha``` : horizontal align
  * ```rotation```

````python
fig, ax = plt.subplots()

fig.suptitle('Figure Title')    # Super Title

ax.plot([1, 3, 2], label='legend')

ax.legend(                      # Legend
    title='This is Legend',
    shadow=True,
    labelspacing=1.2,
    # loc='lower right',
    bbox_to_anchor=[1.2, .5],
    # ncol=2, nrow=2
)

ax.set_title('Ax Title',        # Sub Title
             loc='left', va='bottom',
             fontweight='semibold')        

ax.set_xlabel('X Label')        # X Label
ax.set_ylabel('Y Label')        # Y Label

ax.text(x=1, y=2, s='Text\nis Important',          # 좌표기반 텍스트
        fontsize=20,
        fontweight='bold',
        fontfamily='serif',
        fontstyle='italic',
        color='royalblue',
        linespacing=2,
        backgroundcolor='lightgray',
        alpha=.5,
        va='center',          # vertical align  (top, bottom, center)
        ha='center',          # horizontal align  (left, right, center)
        rotation='vertical',  # vertical, horizontal
        bbox=dict(boxstyle='round', facecolor='wheat', ec='black', alpha=.4)  # bbox
       )            
fig.text(.8, .3, s='Figure Text')    # 기준점 기반 텍스트

plt.show()
````
