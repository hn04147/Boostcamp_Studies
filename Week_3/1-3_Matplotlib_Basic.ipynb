## 그래프 그리기

````python
fig = plt.figure()
ax = fig.add_subplot()

x = np.array([1, 2, 3])

plt.plot(x)
plt.show()
````

````python
fig = plt.figure()

x1 = [1, 2, 3]
x2 = [3, 2, 1]

ax1 = fig.add_subplot(211)
plt.plot(x1)

ax2 = fig.add_subplot(212)
plt.plot(x2)

plt.show()
````

````python
fig = plt.figure()

x1 = [1, 2, 3]
x2 = [3, 2, 1]

ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

ax1.plot(x1)
ax2.plot(x2)

plt.show()
````

````python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1, 1, 1]) # 파랑
ax.plot([1, 2, 3]) # 주황
ax.plot([3, 3, 3]) # 초록
ax.plot([3, 2, 1]) # 빨강

# 동시에 같은 종류의 그래프를 그리면 색상이 자동으로 구분됨
plt.show()
````

````python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1, 2, 3], [3, 4, 5])
ax.bar([1, 2, 3], [1, 2, 3])

# 다른 그래프는 자동으로 색 구분이 안됨
plt.show()
````

````python
# color 파라미터를 사용하여 color 명시 가능
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1, 1, 1], color='r')
ax.plot([2, 2, 2], color='forestgreen')
ax.plot([3, 3, 3], color='#00C3C5')    # rgb hex값 추천

plt.show()
````

````python
# 정보를 추가하기 위해 텍스트 사용 가능
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1, 1, 1], label='1')
ax.plot([2, 2, 2], label='2')
ax.plot([3, 3, 3], label='3')

ax.set_title('Basic Plot')     # 제목 추가
ax.legend()                    # 범례 추가

plt.show()
````

````python
# figure에도 제목 추가 가능
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.plot([1, 2, 3], color='r', label='111')
ax1.plot([3, 2, 1], color='#00C3C5', label='222')

ax2.plot([1, 1, 1], label='1')
ax2.plot([2, 2, 2], label='2')
ax2.plot([3, 3, 3], label='3')

fig.suptitle('FIGURE')         # super title 가장 큰 틀에 있는 타이틀
fig.legend()                   # fig에 범례 지정

ax1.set_title('AX1')
ax2.set_title('AX2')

ax1.legend()
ax2.legend()

plt.show()
````

````python
# .set_{}() 형태의 메서드로 ax에서 특정 데이터를 변경하는 경우
# .get_{}() 형태의 메서드로 해당 정보를 받아올 수 있다.
ax1.get_title()
````

축은 ```ticks```와 ```ticklabels```로 구분한다.  
```ticks```는 축에 적히는 수 위치를 지정한다.

````python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1, 1, 1], label='1')
ax.plot([2, 2, 2], label='2')
ax.plot([3, 3, 3], label='3')

ax.set_title('Basic Plot')

ax.set_xticks([0, 1, 2])    # x축에 적히는 축 설정

ax.legend()
plt.show()
````

````python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1, 1, 1], label='1')
ax.plot([2, 2, 2], label='2')
ax.plot([3, 3, 3], label='3')

ax.set_title('Basic Plot')

ax.set_xticks([0, 1, 2])
ax.set_xticklabels(['one', 'two', 'three'])    # x축에 적히는 텍스트 설정

ax.legend()
plt.show()
````

````python
# 텍스트 추가
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1, 1, 1], label='1')
ax.plot([2, 2, 2], label='2')
ax.plot([3, 3, 3], label='3')

ax.set_title('Basic Plot')

ax.text(x=1, y=2, s='This is text')                # 원하는 위치에 텍스트를 적어줌
ax.annotate(text='This is annotate', xy=(0, 1))    # 원하는 위치에 텍스트를 지정, 포인트를 지정해줌; 화살표 추가 가능

ax.legend()
plt.show()
````

````python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1, 1, 1], label='1')
ax.plot([2, 2, 2], label='2')
ax.plot([3, 3, 3], label='3')

ax.set_title('Basic Plot')

ax.annotate(text='this is annotate', xy=(1, 2),
           xytext=(1.2, 2.2),
           arrowprops=dict(facecolor='black'))      # 텍스트 정보와 x, y 정보를 따로 전달받아서 표현

ax.legend()
plt.show()
````
