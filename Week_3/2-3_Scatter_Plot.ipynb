# Scatter Plot
Scatter plot은 점을 사용하여 두 feature간의 관계를 알기 위하여 사용하는 그래프  
Scatter plot의 목적: 상관 관계 확인 (군집, 값 사이의 차이, 이상치 확인)


점이 많아질수록 점의 분포를 파악하기 힘들다.  
이럴 땐 아래의 방법을 사용한다
* 투명도 조정
* 지터링(jittering): 점의 위치를 약간씩 변경
* 2차원 히스토그램: 히트맵을 사용하여 깔끔한 시각화
* Contour plot(등고선 그래프): 분포를 등고선을 사용하여 표현

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, aspect=1)   # aspect: x축과 y축의 scale을 맞춰주는 역할

np.random.seed(970725)

x = np.random.rand(20)
y = np.random.rand(20)

ax.scatter(x, y)

plt.show()
```

## Scatter Plot의 요소
* 색 (color)
* 모양 (marker)
* 크기 (size)

```python
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, aspect=1)

np.random.seed(970725)

x = np.random.rand(20)
y = np.random.rand(20)
s = np.arange(20) * 20

ax.scatter(x, y,
          s=s,               # Size
          c='white',         # Color
          marker='o',        # Marker
          linewidth=1,
          edgecolor='red')   # 마커의 테두리 색

plt.show()
```

## 정확한 Scatter Plot
#### 붓꽃 데이터셋 실습

```python
iris = pd.read_csv('./iris.csv')
iris.head()
```

```python
iris.describe(include='all')
```

```python
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111)

ax.scatter(x=iris['SepalLengthCm'], y=iris['SepalWidthCm'])

plt.show()
```

```python
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111)

slc_mean = iris['SepalLengthCm'].mean()
swc_mean = iris['SepalWidthCm'].mean()

ax.scatter(x=iris['SepalLengthCm'],
           y=iris['SepalWidthCm'],
           c=['royalblue' if xx <= slc_mean else 'gray' for xx in iris['SepalLengthCm']])

plt.show()
```

```python
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111)

for species in iris['Species'].unique():
    iris_sub = iris[iris['Species']==species]
    ax.scatter(x=iris_sub['SepalLengthCm'],
               y=iris_sub['SepalWidthCm'],
               label=species)
    
ax.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111)

for species in iris['Species'].unique():
    iris_sub = iris[iris['Species']==species]
    ax.scatter(x=iris_sub['PetalLengthCm'],
               y=iris_sub['PetalWidthCm'],
               label=species)
    
ax.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111)

for species in iris['Species'].unique():
    iris_sub = iris[iris['Species']==species]
    ax.scatter(x=iris_sub['PetalLengthCm'],
               y=iris_sub['PetalWidthCm'],
               label=species)

ax.axvline(2.5, color='gray', linestyle=':')
ax.axhline(0.8, color='gray', linestyle=':')
    
ax.legend()
plt.show()
```

```python
fig, axes = plt.subplots(4, 4, figsize=(14, 14))

feat = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

for i, f1 in enumerate(feat):
    for j, f2 in enumerate(feat):
        if i <= j : 
            axes[i][j].set_visible(False)
            continue
        for species in iris['Species'].unique():
            iris_sub = iris[iris['Species']==species]
            axes[i][j].scatter(x=iris_sub[f2], 
                               y=iris_sub[f1], 
                               label=species, 
                               alpha=0.7)
        if i == 3: axes[i][j].set_xlabel(f2)
        if j == 0: axes[i][j].set_ylabel(f1)

plt.tight_layout()        
plt.show()
```
